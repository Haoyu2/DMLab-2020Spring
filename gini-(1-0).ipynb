```python
from gini import *
size = (5, 5)
mat = np.random.choice([1,0],size)
orbi = OrBi(mat)
```

```python
import matplotlib.pyplot as plt

df_giniu_all = orbi.allBi_giniu(base=1.2)
display(df_giniu_all.describe())
df_giniu_all.plot()
display(df_giniu_all.sort_values(by=['Gini_u']))
print()
print("biclusters with lowest gini_u:")
index_min = df_giniu_all.loc[df_giniu_all['Gini_u']==df_giniu_all.min()[0]]
display(index_min)
[display(orbi.show_bi(list(i[0]),list(i[1]))) for i in list(index_min.index)]
print()
print()
print("biclusters with highest gini_u:")
index_max = df_giniu_all.loc[df_giniu_all['Gini_u']==df_giniu_all.max()[0]]
display(index_max)
[display(orbi.show_bi(list(i[0]),list(i[1]))) for i in list(index_max.index)]
print()
```
